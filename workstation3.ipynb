## Advance operation with examples

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 2D Array: Rows = Days (Mon-Fri), Columns = Categories (Tech, Apparel, Home)
sales_data = np.array([
    [2500, 800,  1200], # Monday
    [3100, 650,  1100], # Tuesday
    [2800, 1200, 950 ], # Wednesday
    [4000, 900,  1300], # Thursday
    [3500, 1100, 1500]  # Friday
])
print("Sales Data Array:")
print(sales_data)
print("\n ====sales analysis ====")
print("sales data shape:", sales_data.shape)
```

```python
# Total Sales per Category (Column-wise sum)
category_totals = sales_data.sum(axis=0)
print("category totals:" , category_totals)

# Average Daily Sales (Row-wise mean)
daily_averages = sales_data.mean(axis=1)
print("daily average:\n", daily_averages)

# Maximum Sale in 'Tech' (First column)
max_tech = np.max(sales_data[:, 0])
print("maximum sale in tech category:", max_tech)

max_sales= np.max(sales_data[:,:])
print("max sales in all category:", max_sales)

Tax_cal= np.array(0.08 * sales_data)
print("Tax Calculation:\n", Tax_cal)

print("sales more the 1500:\n", np.array(sales_data>=1500,dtype=int))
```

```python
cumulative_sales= np.cumsum(sales_data,axis=1)
print("sales_data:\n", sales_data)
print("cumulative_sales:\n", cumulative_sales)

plt.figure(figsize=(10,6))
plt.plot(np.mean(cumulative_sales,axis=0))
plt.title("average cumulative sales across all category")
plt.xlabel("weekdays")
plt.ylabel("sales")
plt.show()
```
