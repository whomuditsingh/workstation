### Numpy Array

```python
import numpy as np
```

creating an array from a list

```python
array_1=np.array([40,42,43])
print("1D array:",array_1)
array_2= np.array([[21,22,23],[6,7,8]])
print("2D array:",array_2)
```

```python
list_1=[1,2,3]
print("multiplication on list", list_1*2)

array_3= np.array([1,2,3])
print("multiplication on array:", array_3*2)

import time
start= time.time()
list_1=[i*2 for i in range(1000000)]
print("\n list operation time:", time.time()-start)

start= time.time()
np_array = np.arange(1000000)
print("\n Numpy operation time:", time.time()- start)
```

### creating an array from scratch

```python
zeroes= np.zeros((3,4))
print("zeros array:\n", zeroes)

ones= np.ones((3,4))
print("ones array: \n", ones)

full_method= np.full((3,4),7)
print("full array:\n", full_method)

random= np.random.random((2,2))
print("random array:\n", random)

sequence= np.arange(0,11,2)
print("sequence array:", sequence)

```

## vector,matrix and tensor

```python
vector= np.array([1,2,3])
print("vector array:", vector)

matrix= np.array([[1,2,3],[3,4,5]])
print("matrix array:\n", matrix)

tensor= np.array([[[1,2],[3,4]],[[5,6],[7,8]]])
print("tensor array:\n", tensor)


```
